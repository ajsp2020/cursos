```python
# Debugandoo com PDB

### PDB -> Python Debugger


#### A utilização do print() para deguggar código é uma prática ruim.
```

#### Comandos básicos doo PDB

#### l (listar onde estamos no código)
#### n (próoxima linha)
#### p (imprime variável)
#### c (Continua a execução - finaliza o debbuggin)


```python
import pdb

nome = 'Angelina'
sobrenome = 'Jolie'
pdb.set_trace()
nome_completo = nome + ' ' + sobrenome
curso = 'Ciencias da computação'
final = nome_completo + ' faz o curso ' + curso

print(final)
```
