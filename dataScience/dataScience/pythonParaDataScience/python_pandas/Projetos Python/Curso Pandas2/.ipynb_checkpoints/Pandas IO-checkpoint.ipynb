# Criando nomes

```python
import pandas as pd
```

```python
nomes_m = pd.read_json("https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=m")
nomes_f = pd.read_json("https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=f")
```

```python
print("Quantidade de nomes: " + str(len(nomes_f) + len(nomes_m)))
```

```python
frames = [nomes_f, nomes_m]
```

```python
frames
```

```python
nomes = pd.concat(frames)['nome'].to_frame() # Concatenando os DataFrames, aplicando a uma coluna so e convertendo para um dataframe
```

```python
nomes.sample(5) # .sample -> pega um nome aleatório
```

# Incluindo os ID dos alunos

```python
import numpy as np
np.random.seed(123)
```

```python
total_alunos = len(nomes) # Qtdade de alunos
total_alunos
```

```python
# Criando uma nova coluna 
nomes["id_aluno"] = np.random.permutation(total_alunos) + 1 # distribui os alunos randomicamente e aplica eles de forma aleatórias
```

```python
nomes.sample(5)
```

```python
# Criando 2 domínios de email
dominios = ['@diminiodiemail.com.br', '@servicodoemail.com']
nomes['dominio'] = np.random.choice(dominios, total_alunos) # choice para poder escolher algum elemento da nossa lista de domínio
```

```python
nomes.sample(5)
```

```python
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower() # foi criando uma nova coluna com o email depois pegou a coluna nome transformou a coluna dominio em str e concatenou depois pegou o resultado e deixou em letra minuscula
```

```python
nomes.sample(5)
```

# Selecionado cursos

```python
# saber quantas matriculas cada aluno tem
todas_matriculas = [] # criando uma array vazia
x = np.random.rand(20) # calcular de maneira randomica os 20 curso
prob = x / sum(x) # probabilidade
```

```python
np.random.rand(20)
```

```python
po = np.random.rand(20) / sum(np.random.rand(20))
po
```

```python
nomes.head()
```

```python
for index, row in nomes.iterrows(): # iterrows -> funçao que passa linha a linha
    id= row.id_aluno # saber qual a id do aluno
    matriculas = row.matriculas # saber a quantidade de matriculas desse aluno
    for i in range(matriculas): # saber quantos cursos preciso escolher para esse aluno baseado na quantidade de matricula
        mat = [id, np.random.choice(cursos.index, p = prob)] # vou escolher uma matricula
        todas_matriculas.append(mat)
        
matriculas = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])
```

```python
matriculas.head()
```

```python
matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'id_aluno':'quatidade_de_alunos'})
```

```python
nomes.sample(10)
```

```python
cursos.head()
```

```python
matriculas.head()
```

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'id_aluno':'quantidade_de_alunos'})
```
